```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
import re
import string
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import plot_confusion_matrix
from sklearn.svm import SVC, NuSVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from scipy.stats import norm
```

```python
train_spectral = pd.read_csv(r'C:\Users\Callum\Documents\cab420-assignment1a\CAB420Assignment1A\training.csv')
test_spectral = pd.read_csv(r'C:\Users\Callum\Documents\cab420-assignment1a\CAB420Assignment1A\testing.csv')
```

```python
test_spectral['class'] = pd.Categorical(test_spectral['class'])
train_spectral['class'] = pd.Categorical(train_spectral['class'])

predictionColumn_test = test_spectral['class']
predictors_test = test_spectral.drop('class', axis=1)

Y_train = train_spectral['class']
X_train = train_spectral.drop('class', axis=1)

X_test, X_val, Y_test, Y_val = train_test_split(predictors_test, predictionColumn_test, test_size=0.9, random_state=1)


```

```python
import matplotlib.pyplot as plt
spectral_hist = plt.hist(train_spectral['class']) 
plt.title("Histogram with spectral")
plt.show()
```

Good spread

```python
def eval_model(model, X_train, Y_train, X_test, Y_test, graphs):
    if(graphs):
        fig = plt.figure(figsize=[25, 8])
        ax = fig.add_subplot(1, 2, 1)
        conf = plot_confusion_matrix(model, X_train, Y_train, normalize='true', ax=ax)
        conf.ax_.set_title('Training Set Performance');
        ax = fig.add_subplot(1, 2, 2)
        conf = plot_confusion_matrix(model, X_test, Y_test, normalize='true', ax=ax)
        conf.ax_.set_title('Validation Test Set Performance');
    pred = model.predict(X_test)
    print(classification_report(Y_test, pred)) 
    return str(sum(pred == Y_test)/len(Y_test))
```

```python
#Make the model
cknn = KNeighborsClassifier(n_neighbors=9, weights='distance')
cknn.fit(X_train, Y_train)
test_accur = eval_model(cknn, X_train, Y_train, X_val, Y_val, True)
print("Test Accuracy: ", test_accur)
```

Not too bad, this model is using distance weight and n_neighbors is set at 10.
We can try to optimize this to find the best combination.

```python
test_accuracy = {}
for i in range(3, 25, 2):
    for x in range(0,2):
        if(x == 0):
            cknn = KNeighborsClassifier(n_neighbors=i, weights='distance')
        else:
            cknn = KNeighborsClassifier(n_neighbors=i)
        cknn.fit(X_train, Y_train)
        test_accur = eval_model(cknn, X_train, Y_train, X_val, Y_val, False)
        if(x == 0):
            test_accuracy[i, "Dist"] = (test_accur)
        else:
            test_accuracy[i, "No Dist"] = (test_accur)

print(test_accuracy)
import operator
max(test_accuracy.items(), key=operator.itemgetter(1))[0]
```

Here we see that using the n_neighbors value of 3 is the most optimal for always ascribing the data to the correct group. The distance variable makes no difference to the accuracy of the model to due to the n_neighbors value being low. Here is the confusion matrix for 3.

```python
#Make the model
cknn = KNeighborsClassifier(n_neighbors=3)
cknn.fit(X_train, Y_train)
test_accur = eval_model(cknn, X_train, Y_train, X_val, Y_val, True)
print("Test Accuracy: ", test_accur)
```

```python
#This is the automated way of doing this
from sklearn.model_selection import GridSearchCV
#create new a knn model
knn2 = KNeighborsClassifier()
#create a dictionary of all values we want to test for n_neighbors
param_grid = {'n_neighbors': np.arange(3, 25), 'weights': ['distance', 'uniform']}
#use gridsearch to test all values for n_neighbors
knn_gscv = GridSearchCV(knn2, param_grid, cv=5)
#fit model to data
knn_gscv.fit(X_train, Y_train)
#check top performing n_neighbors value

predictions = knn_gscv.predict(X_test) 
print(classification_report(Y_test, predictions)) 

knn_gscv.best_params_
```

This automated way produces the same result

```python
estimator__kernel = ['rbf']
gamma_values = ['scale', 'auto']
accuracy_max = 0
accur_obj = {}
for i in range(1,50):
     for y,gammas in enumerate(gamma_values):
        for x,elem in enumerate(estimator__kernel):
            search = SVC(class_weight='balanced', C=i, kernel=elem, gamma = gammas)
            search.fit(X_train, Y_train)
            predicted = search.predict(X_test)
            accur = accuracy_score(Y_test, predicted)
            if(accur > accuracy_max):
                best = search
                accuracy_max = accur
                name = "C:" , i+1, " Kernel: ", elem , " Gamma: ", gammas
                accur_obj = {name: accuracy_max}
eval_model(best, X_train, Y_train, X_test, Y_test, True)


print(accur_obj)
```

Lets optimize the hyperparamters for 1vall

```python
estimator__kernel = ['rbf']
gamma_values = ['scale', 'auto']
accuracy_max = 0
accur_obj = {}
for i in range(1,50):
    for y,gammas in enumerate(gamma_values):
        for x,elem in enumerate(estimator__kernel):
            search = OneVsRestClassifier(SVC(class_weight='balanced', C=i, kernel=elem, gamma=gammas))
            search.fit(X_train, Y_train)
            predicted = search.predict(X_test)
            accur = accuracy_score(Y_test, predicted)
            if(accur > accuracy_max):
                best2 = search
                accuracy_max = accur
                name = "C:" , i+1, " Kernel: ", elem , " Gamma: ", gammas
                accur_obj = {name: accuracy_max}
eval_model(best2, X_train, Y_train, X_test, Y_test, True)


print(accur_obj)

```
